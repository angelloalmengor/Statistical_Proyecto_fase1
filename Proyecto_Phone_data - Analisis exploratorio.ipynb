```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
phone =pd.read_csv("phone_data.csv")
phone.head()
```

### Ánalisis Exploratorio Dataset Original

```python
# Mostrar las primeras filas del dataset
print(phone.head())
```

```python
# Mostrar información sobre el dataset
print(phone.info())
```

```python
# Describir estadísticamente el dataset
print(phone.describe())
```

```python
# Verificar valores nulos
print(phone.isnull().sum())
```

```python
#Porcentaje faltantes por columna
phone.isnull().mean()
```

```python
#Tipo de variable
phone.dtypes
```

```python
# Histograma de la distribución de precios
plt.figure(figsize=(10, 6))
sns.histplot(phone['best_price'], bins=20, kde=True)
plt.title('Distribución de precios de los teléfonos')
plt.xlabel('Mejor precio')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Diagrama de dispersión de precio vs. capacidad de la batería
plt.figure(figsize=(10, 6))
sns.scatterplot(x='battery_size', y='best_price', data=phone)
plt.title('Precio vs. Capacidad de la batería')
plt.xlabel('Capacidad de la batería')
plt.ylabel('best_price')
plt.show()
```

```python
# Boxplot de la resolución de la pantalla por marca
plt.figure(figsize=(12, 8))
sns.boxplot(x='os', y='brand_name', data=phone)
plt.title('Marca de dispositivo por sistema operativo')
plt.xlabel('Sistema operativo')
plt.ylabel('Marca de dispositivo')
plt.xticks(rotation=45)
plt.show()
```
