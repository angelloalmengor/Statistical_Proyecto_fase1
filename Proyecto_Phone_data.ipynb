## Ingenieria de características

```python
# Clasificaremos entre variables categóricas, continuas y discretas
categoricas = [col for col in phone.columns if(phone[col].dtypes == 'object')]
categoricas
```

```python
# Clasificaremos entre variables categóricas, continuas y discretas
continuas = [col for col in phone.columns if((phone[col].dtypes in ['int64', 'float64']) and len(phone[col].unique()) > 30)]
continuas
```

```python
# Clasificaremos entre variables categóricas, continuas y discretas
discretas = [col for col in phone.columns if((phone[col].dtypes in ['int64', 'float64']) and len(phone[col].unique()) <= 30)]
discretas
```

```python
#imputación de variables continuas
continuas_con_na = [col for col in continuas if phone[col].isnull().mean() > 0]
phone[continuas_con_na].isnull().mean() 
```

##### Imputación por Media y Mediana
Variable lowest_price

```python
# Imputación por Media
mean_lowest_price = round(phone['lowest_price' ].mean(), 2)

temp_series = phone['lowest_price'].fillna(mean_lowest_price)

fig = plt.figure(figsize=(5, 3))
temp_series.hist(bins=20, density=True, color='red', alpha=0.3)
plt.title('lowest_price - ' + str(mean_lowest_price))
plt.show()
```

```python
##Imputacion por mediana
median_lowest_price = round(phone['lowest_price'].median(), 2)

temp_series = phone['lowest_price'].fillna(median_lowest_price )

fig = plt.figure(figsize=(5, 3))
temp_series.hist(bins=20, density=True, color='red', alpha=0.3)
plt.title('lowest_price - ' + str(median_lowest_price ))
plt.show()
```

```python
#Imputacion
phone.fillna({'lowest_price': 'median_lowest_price'}, inplace=True)
```

Variable highest_price

```python
# Imputación por Media
mean_highest_price = round(phone['highest_price' ].mean(), 2)

temp_series = phone['lowest_price'].fillna(mean_highest_price)

fig = plt.figure(figsize=(5, 3))
temp_series.hist(bins=20, density=True, color='red', alpha=0.3)
plt.title('highest_price - ' + str(mean_highest_price))
plt.show()
```

```python
##Imputacion por mediana
median_highest_price = round(phone['highest_price'].median(), 2)

temp_series = phone['highest_price'].fillna(median_highest_price )

fig = plt.figure(figsize=(5, 3))
temp_series.hist(bins=20, density=True, color='red', alpha=0.3)
plt.title('highest_price - ' + str(median_highest_price ))
plt.show()
```

```python
#Imputacion
phone.fillna({'highest_price': 'median_highest_price'}, inplace=True)
```

Variable battery_size

```python
# Imputación por Media
mean_battery_size = round(phone['battery_size' ].mean(), 2)

temp_series = phone['battery_size'].fillna(mean_battery_size)

fig = plt.figure(figsize=(5, 3))
temp_series.hist(bins=20, density=True, color='red', alpha=0.3)
plt.title('battery_size - ' + str(mean_battery_size))
plt.show()
```

Variables tratamiento de Outliers

```python
phone_proy =pd.read_csv("fase_1_proy.csv")
phone_proy.head()
```

```python
def get_variables_scale(phone_proy):
    continuas = [col for col in phone_proy.columns if phone_proy[col].dtype in ['float64','int64'] and len(phone_proy[col].unique())>30]
    discretas = [col for col in phone_proy.columns if phone_proy[col].dtype in ['float64','int64'] and len(phone_proy[col].unique())<=30]

    return continuas, discretas
```

```python
#Aplicacion de variables
continuas, discretas = get_variables_scale(phone_proy)
continuas
```

```python
# creamos un dataframe para las variables continuas 
continuas_proy = pd.DataFrame(phone_proy)
continuas, discretas = get_variables_scale(phone_proy)
df_continuas =continuas_proy [continuas]
df_continuas.head()
```
